## Creating new columns

```python
# Create column
df['ExpLevel'] = 'Entry level'

# Update based on boolean mask
mask = df['TotalWorkingYears'] > 5
df.loc[mask, 'ExpLevel'] = 'Mid level'

mask = df['TotalWorkingYears'] > 12
df.loc[mask, 'ExpLevel'] = 'Senior level'

mask = df['TotalWorkingYears'] > 25
df.loc[mask, 'ExpLevel'] = 'Advanced/Executive'
```

```python
manager_role = ['Manager']
df.loc[df['JobRole'].isin(manager_role), 'IsManager'] = 'True'

df.loc[~df['JobRole'].isin(manager_role), 'IsManager'] = 'False'

df['IsManager'].value_counts()
```

```python
# Ensure the mapping covers all unique levels
satisfaction_mapping = {1: 'Low', 2: 'Medium-Low', 3: 'Medium-High', 4: 'High'}

# Group and calculate percentage
groupby = df.groupby(['JobSatisfaction', 'Attrition'])['MonthlyIncome'].count().unstack(1)
percentage_calc = groupby['Yes'] / (groupby['Yes'] + groupby['No']) * 100

# Apply the mapping
percentage_calc.index = percentage_calc.index.map(satisfaction_mapping)

# Plot
percentage_calc.plot(kind='bar', color=sns.color_palette("rocket"), rot=0)
plt.xlabel('Job Satisfaction')
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Job Satisfaction')
```

```python
# Mapping for Environment Satisfaction levels
environment_satisfaction_mapping = {1: 'Low', 2: 'Medium-Low', 3: 'Medium-High', 4: 'High'}

# Group and calculate percentage
groupby_env = df.groupby(['EnvironmentSatisfaction', 'Attrition'])['MonthlyIncome'].count().unstack(1)
percentage_calc_env = groupby_env['Yes'] / (groupby_env['Yes'] + groupby_env['No']) * 100

# Apply the mapping
percentage_calc_env.index = percentage_calc_env.index.map(environment_satisfaction_mapping)

# Plot
percentage_calc_env.plot(kind='bar', color=sns.color_palette("rocket"), rot=0)

# Set labels and title
plt.xlabel('Environment Satisfaction')
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Environment Satisfaction')
```

```python
# Mapping for Work Life Balance levels
wlb_mapping = {1: 'Low', 2: 'Medium-Low', 3: 'Medium-High', 4: 'High'}

# Group and calculate percentage
groupby_wlb = df.groupby(['WorkLifeBalance', 'Attrition'])['MonthlyIncome'].count().unstack(1)
percentage_calc_wlb = groupby_wlb['Yes'] / (groupby_wlb['Yes'] + groupby_wlb['No']) * 100

# Apply the mapping
percentage_calc_wlb.index = percentage_calc_wlb.index.map(wlb_mapping)

# Plot
percentage_calc_wlb.plot(kind='bar', color=sns.color_palette("rocket"), rot=0)

# Set labels and title
plt.xlabel('Work Life Balance')
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Work Life Balance')
```

```python
df_grouped = df.groupby(['Attrition','ExpLevel'])['MonthlyIncome'].mean().unstack(0)

# Order index as most experienced -> least
order = ['Entry level','Mid level','Senior level','Advanced/Executive']
df_grouped = df_grouped.reindex(order)

# Plot
ax = df_grouped.plot(kind='barh',color=['royalblue','orange'], xlabel='Monthly Income', figsize=(8,5))
plt.ylabel('Expirience Level')
plt.title("Attrition Impact on Monthly Income Across Experience Levels")
```

```python
df_grouped = df.groupby(['Attrition','ExpLevel'])['MonthlyIncome'].mean().unstack(0)

# Order index as most experienced -> least
order = ['Entry level','Mid level','Senior level','Advanced/Executive']
df_grouped = df_grouped.reindex(order)

# Subtract attrition vs non-attrition income
income_diff = df_grouped['No'] - df_grouped['Yes']

# Plot bar chart
income_diff.plot(kind='bar', color='royalblue',rot=0)

plt.xlabel('Experience Level')
plt.ylabel('Difference in Average Income')
plt.title('Income Difference Attrition vs Non-Attrition Workers')
```

```python
managers = df[df['IsManager'] == 'True']
non_managers = df[df['IsManager'] == 'False']

manager_attrition_rate = len(managers[managers['Attrition']=='Yes']) / len(managers) * 100 if len(managers) > 0 else 0
non_manager_attrition_rate = len(non_managers[non_managers['Attrition']=='Yes']) / len(non_managers) * 100 if len(non_managers) > 0 else 0

categories = ['Managers', 'Non-Managers']
rates = [manager_attrition_rate, non_manager_attrition_rate]

# Creating the bar plot
#plt.figure(figsize=(8, 5))
plt.bar(categories, rates, color=['royalblue','orange'])

# Adding title and labels
plt.title('Attrition Rates: Managers vs Non-Managers')
plt.xlabel('Category')
plt.ylabel('Attrition Rate (%)')
```

```python
# Group by 'Attrition' and 'EducationField', then calculate the mean 'MonthlyIncome'
grouped_df = df.groupby(['Attrition', 'EducationField'])['MonthlyIncome'].median().unstack(0)

# Calculate the difference between Non-Attrition and Attrition Monthly Incomes
grouped_df['Income_Difference'] = grouped_df['No'] - grouped_df['Yes']

# Sort the DataFrame by this difference
sorted_df = grouped_df.sort_values('Income_Difference', ascending=True)

# Plot the sorted DataFrame without the 'Income_Difference' column
sorted_df.drop(columns='Income_Difference').plot(kind='barh')

# Set plot title and labels
plt.title('Median Monthly Income by Education Field: Attrition vs Non-Attrition')
plt.xlabel('Median Monthly Income')
plt.ylabel('Education Field')
```

```python
grouped_df = df.groupby(['Attrition', 'EducationField'])['MonthlyIncome'].median().unstack(0)

# Calculate the difference in median incomes
grouped_df['Difference'] = grouped_df['No'] - grouped_df['Yes']

# Sort the DataFrame by the 'Difference' column
sorted_df = grouped_df.sort_values('Difference')

# Create a colormap
cmap = sns.color_palette("crest", as_cmap=True)

# Normalize the 'Difference' values for color mapping
norm = plt.Normalize(sorted_df['Difference'].min(), sorted_df['Difference'].max())

# Plot
fig, ax = plt.subplots()
bars = ax.barh(sorted_df.index, sorted_df['Difference'], color=cmap(norm(sorted_df['Difference'].values)))

# Set plot title and labels
#plt.title('Difference in Median Monthly Income: Attrition vs Non-Attrition')
plt.xlabel('Difference in Median Monthly Income')
plt.ylabel('Education Field')
```
